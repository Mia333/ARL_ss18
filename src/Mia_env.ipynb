```python
# env.py
######################################
#     Build walk environment         #
######################################

import pyglet
import numpy as np
from math import *

class WalkEnv(object):
    viewer = None       # 首先没有 viewer

    def __init__(self):
        pass

    def step(self, action):
        pass

    def reset(self):
        pass
    
    def render(self):
        if self.viewer is None: # 如果调用了 render, 而且没有 viewer, 就生成一个
            self.viewer = Viewer()
        self.viewer.render()    # 使用 Viewer 中的 render 功能



class Viewer(pyglet.window.Window):
    bar_thc = 3     # 手臂的厚度
    
    def __init__(self):
        # 画出Poppy
        # 创建窗口的继承
        # vsync 如果是 True, 按屏幕频率刷新, 反之不按那个频率
                # vsync=False to not use the monitor FPS, we can speed up training
        super(Viewer, self).__init__(width=400, height=400, resizable=False, caption='Walkgesture', vsync=False)
        
        # 窗口背景颜色
        pyglet.gl.glClearColor(1, 1, 1, 1)
        
        # 将手臂的作图信息放入这个 batch
        self.batch = pyglet.graphics.Batch()    # display whole batch at once
        
        ## Left_Leg　
        #len: 50, width: 10
        self.leg_upper_l = self.batch.add(
            4, pyglet.gl.GL_QUADS, None,
            ('v2f', [250, 150,              # coordinate
                     230, 200,
                     240, 200,
                     260, 150]),
            ('c3B', (249, 86, 86) * 4,))    # color
        
        self.leg_lower_l = self.batch.add(
            4, pyglet.gl.GL_QUADS, None,
            ('v2f', [250, 150,              
                     260, 150,
                     260, 100,
                     250, 100]),
            ('c3B', (249, 86, 86) * 4,))    # color
        ## Right_Leg　
        # len: 50, width: 10
        self.leg_upper_r = self.batch.add(
            4, pyglet.gl.GL_QUADS, None,
            ('v2f', [245, 150,              # coordinate
                     225, 200,
                     235, 200,
                     255, 150]),
            ('c3B', (86, 86, 86) * 4,))    # color
        
        self.leg_lower_r = self.batch.add(
            4, pyglet.gl.GL_QUADS, None,
            ('v2f', [245, 150,              
                     255, 150,
                     255, 100,
                     245, 100]),
            ('c3B', (86, 86, 86) * 4,))    # color
        
        ## Body
        self.body = self.batch.add(
            4, pyglet.gl.GL_QUADS, None,
            ('v2f', [200, 200,              
                     260, 200,
                     245, 270,
                     215, 270]),
            ('c3B', (0,0,255)*4,))
        
        ## Head
        #x = 230
        #y = 275
        #r = 5
        #p = 8
        #deg = 360 / p
        #P = x, y # P for POINTS
        #for i in range(p):
        #    n = deg * i
        #    P += int(r * cos(n)) + x, int(r * sin(n)) + y
        #self.head = self.batch.add(p+1, pyglet.gl.GL_TRIANGLE_FAN, None, ('v2i', P), ('c3B', (0,0,255),))
        
        def makeCircle(numPoints):
            verts = []
            for i in range(numPoints):
                angle = radians(float(i)/numPoints * 360.0)
                x = 15*cos(angle) + 230
                y = 15*sin(angle) + 285
                verts += [x,y]
            return verts
        numPoints = 100
        verts = makeCircle(numPoints)
        verts = [int(i) for i in verts]
        self.head = self.batch.add(
            numPoints, pyglet.gl.GL_POINTS, None, 
            ('v2i', verts), 
            ('c3B', (1,1,255)*numPoints,))
        
            
    def render(self):
        # 刷新并呈现在屏幕上
        self._update_leg()  # 更新腿内容 (暂时没有变化)
        self.switch_to()
        self.dispatch_events()
        self.dispatch_event('on_draw')
        self.flip()
    def on_draw(self):
        # 刷新腿等位置
        self.clear()        # 清屏
        self.batch.draw()   # 画上 batch 里面的内容
    def _update_leg(self):
        # 更新手臂的位置信息
        pass
```
